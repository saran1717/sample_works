```python
import pandas as pd
import numpy as np
```

```python
na_vals=['NA','Missing']
people={
        'first':['anand','arun','janu','akhil',np.nan,None,'NA'],
        'last':['bhanu','rajesh','rajesh','rajan',np.nan,np.nan,'Missing'],
        'email':['anandpvr@gmail.com','bhanuj@gmail.com','arjunpvr@gmail.com',None,np.nan,'Anonymous@email.com','NA'],
       'age':['33','53','43','63',None,None,'Missing']
       }
```

```python
df=pd.DataFrame(people)
df.replace('NA',np.nan,inplace=True)
df.replace('Missing',np.nan,inplace=True)
```
